```python
import pandas as pd
```

```python
df=pd.read_csv("C:/python/Jadi_ML/01 Iman/3-1 heart.csv" ,index_col=False)
x=df[['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh','exng', 'oldpeak', 'slp', 'caa', 'thall']].values
y=df['output'].values
df
```

Train/Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

Model Training

```python
from sklearn.tree import DecisionTreeClassifier
htree=DecisionTreeClassifier(criterion="entropy")
htree.fit(x_train,y_train)

yhat=htree.predict(x_test)
```

Evaluation

```python
from sklearn import metrics
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(yhat, y_test))
```

```python
print(metrics.classification_report(yhat,y_test))
print(metrics.confusion_matrix(yhat,y_test))
```
